1. What is Simple Linear Regression (SLR)? Explain its purpose.

It is a regression technique that models the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.
Its purpose is to predict continuous values and understand how changes in X affect Y.

Example: Predicting house price (Y) based on house size in square feet (X).

2. What are the key assumptions of Simple Linear Regression?

There is a linear relationship between the independent variable (X) and the dependent variable (Y).
The errors (residuals) are independent, normally distributed, and have constant variance.

Example: Predicting salary based on years of experience assumes salary increases linearly with experience.

3. Write the mathematical equation for a simple linear regression model and
explain each term.

The equation is Y = mX + c, where m is the slope (rate of change) and c is the intercept (value of Y when X = 0).
X is the independent variable and Y is the dependent variable being predicted.

Example: If m = 2 and c = 5, then for X = 3, Y = 11.

4. Provide a real-world example where simple linear regression can be
applied.

It is used when one factor directly influences another in a measurable, linear way.
It helps in predicting future values based on past data.

Example: Predicting electricity bill amount based on monthly units consumed.

5. What is the method of least squares in linear regression?

It finds the best-fit line by minimizing the sum of squared differences between actual and predicted values.
It ensures the regression line represents the data with minimum overall error.

Example: Fitting a line that minimizes squared errors between actual and predicted house prices.

6. What is Logistic Regression? How does it differ from Linear Regression?

It is a supervised learning algorithm used for binary classification that predicts probabilities using a sigmoid function.
Unlike Linear Regression which predicts continuous values, Logistic Regression predicts class labels (0 or 1).

Example: Predicting whether an email is spam (1) or not spam (0).

7. Name and briefly describe three common evaluation metrics for regression
models.

MAE measures the average absolute difference between actual and predicted values, while MSE squares errors to penalize larger mistakes.
RMSE is the square root of MSE and shows error in the same unit as the target variable.

Example: Using RMSE to measure how close predicted house prices are to actual prices

8.  What is the purpose of the R-squared metric in regression analysis?

It measures how well the regression model explains the variance in the dependent variable.
It helps in comparing models, where a higher R² indicates a better fit.

Example: An R² of 0.90 means the model explains 90% of the variation in house prices.

9. Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# sample data
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

# fit model
model = LinearRegression()
model.fit(X, y)

# print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

10. How do you interpret the coefficients in a simple linear regression model?

The slope (m) shows how much the dependent variable (Y) changes for a one-unit increase in X.
The intercept (c) represents the value of Y when X = 0.

Example: If slope = 3, salary increases by ₹3,000 for each additional year of experience.
